```python
import pandas as pd
weather_features = ['Max Temp (°C)', 'Min Temp (°C)', 'Mean Temp (°C)', 'Heat Deg Days (°C)',
                    'Cool Deg Days (°C)', 'Total Precip (mm)', 'Snow on Grnd (cm)']

# Grouping by 'Year' and 'Month' and calculating the mean values for weather features
yearmonth_weather_avg = winnipegcol_df.groupby(['Year','Month'])[weather_features].mean()

# Displaying the calculated averages
yearmonth_weather_avg
```

```python


import matplotlib.pyplot as plt

# Resetting the index to have 'Year' and 'Month' as regular columns
yearmonth_weather_avg_reset = yearmonth_weather_avg.reset_index()

# Plotting the 'Max Temp (°C)' column
plt.figure(figsize=(8, 4))
for year in yearmonth_weather_avg_reset['Year'].unique():
    year_data = yearmonth_weather_avg_reset[yearmonth_weather_avg_reset['Year'] == year]
    plt.plot(year_data['Month'], year_data['Max Temp (°C)'], label=str(year))

plt.title('Mean Max Temp (°C)')
plt.xlabel('Month')
plt.ylabel('Max Temp (°C)')
plt.legend()
plt.xticks(yearmonth_weather_avg_reset['Month'])  # Set x-axis ticks to the unique months
plt.tight_layout()
plt.show()



# Plotting the 'Min Temp (°C)' column
plt.figure(figsize=(8, 4))
for year in yearmonth_weather_avg_reset['Year'].unique():
    year_data = yearmonth_weather_avg_reset[yearmonth_weather_avg_reset['Year'] == year]
    plt.plot(year_data['Month'], year_data['Min Temp (°C)'], label=str(year))

plt.title('Mean Min Temp (°C)')
plt.xlabel('Month')
plt.ylabel('Min Temp (°C)')
plt.legend()
plt.xticks(yearmonth_weather_avg_reset['Month'])  # Set x-axis ticks to the unique months
plt.tight_layout()
plt.show()

# Plotting the 'Heat Deg Days (°C)' column
plt.figure(figsize=(8, 4))
for year in yearmonth_weather_avg_reset['Year'].unique():
    year_data = yearmonth_weather_avg_reset[yearmonth_weather_avg_reset['Year'] == year]
    plt.plot(year_data['Month'], year_data['Heat Deg Days (°C)'], label=str(year))

plt.title('Mean Heat Deg Days (°C)')
plt.xlabel('Month')
plt.ylabel('Heat Deg Days (°C)')
plt.legend()
plt.xticks(yearmonth_weather_avg_reset['Month'])  # Set x-axis ticks to the unique months
plt.tight_layout()
plt.show()

# Plotting the 'Cool Deg Days (°C)' column
plt.figure(figsize=(8, 4))
for year in yearmonth_weather_avg_reset['Year'].unique():
    year_data = yearmonth_weather_avg_reset[yearmonth_weather_avg_reset['Year'] == year]
    plt.plot(year_data['Month'], year_data['Cool Deg Days (°C)'], label=str(year))

plt.title('Cool Deg Days (°C)')
plt.xlabel('Month')
plt.ylabel('Cool Deg Days (°C)')
plt.legend()
plt.xticks(yearmonth_weather_avg_reset['Month'])  # Set x-axis ticks to the unique months
plt.tight_layout()
plt.show()

# Plotting the 'Total Precip (mm)' column
plt.figure(figsize=(8, 4))
for year in yearmonth_weather_avg_reset['Year'].unique():
    year_data = yearmonth_weather_avg_reset[yearmonth_weather_avg_reset['Year'] == year]
    plt.plot(year_data['Month'], year_data['Total Precip (mm)'], label=str(year))

plt.title('Total Precip (mm)')
plt.xlabel('Month')
plt.ylabel('Total Precip (mm)')
plt.legend()
plt.xticks(yearmonth_weather_avg_reset['Month'])  # Set x-axis ticks to the unique months
plt.tight_layout()
plt.show()


# Plotting the 'Snow on Grnd (cm)' column
plt.figure(figsize=(8, 4))
for year in yearmonth_weather_avg_reset['Year'].unique():
    year_data = yearmonth_weather_avg_reset[yearmonth_weather_avg_reset['Year'] == year]
    plt.plot(year_data['Month'], year_data['Snow on Grnd (cm)'], label=str(year))

plt.title('Snow on Grnd (cm)')
plt.xlabel('Month')
plt.ylabel('Snow on Grnd (cm)')
plt.legend()
plt.xticks(yearmonth_weather_avg_reset['Month'])  # Set x-axis ticks to the unique months
plt.tight_layout()
plt.show()


```
